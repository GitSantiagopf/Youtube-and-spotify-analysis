```python
#import libraries to use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the database
data = pd.read_csv('Spotify_Youtube.csv')
data.head()
```

```python
#Variable information 
data.info()
```

```python
#Delete categories
data = data.drop(['Title', 'Url_spotify', 'Uri', 'Description', 'official_video', 'Channel', 'Url_youtube'], axis = 1)
```

**our dataset by album types to represnt share of views of each type**




```python
#Filter album type by views
streamperalbum=data.groupby('Album_type').sum().loc[:,['Views','Stream','Likes','Comments']]
```

```python
streamperalbum.plot.pie(y='Views',colors=["#991E1E", "#A54545","#EE7575"],autopct='%1.0f%%').legend(bbox_to_anchor=(1.5, 1))
plt.title('Views Per Album Type')
plt.show()
```

*The pie chart represents the distribution of different types of music releases, with albums contributing to the majority of the releases at 74.4%. Singles represent 22.1% and compilations represent the remaining 3.5%. This indicates that albums are the most popular form of music publication, followed by singles and compilations.*

Group our dataset by album types to represnt share of stream of each type


```python
#Filter album type by streams
streamperalbum.plot.pie(y='Stream',colors=["#1E9941", "#44C167","#68F18F"],autopct='%1.0f%%').legend(bbox_to_anchor=(1.5, 1))
plt.title('Streams Per Album Type')
plt.show()
```

*The pie chart shows the proportion of different types of music releases, with albums representing the largest portion at 80%. In second place are singles, at 18%, while compilations occupy the remaining 2%. This suggests that albums are the most commonly used medium for releasing music, followed by singles and compilations.*


**Most viewed artists on Youtube platform**

```python
#Filter top 3 artist on youtube
artists_youtube = data.groupby('Artist').sum()['Views'].sort_values(ascending=False).head(5)
artists_youtube = artists_youtube.sort_values(ascending = False)
artists_youtube = pd.DataFrame(artists_youtube)
```

```python
artists_youtube.reset_index()["Artist"]
```

```python
plt.figure(figsize=(19,6))
plt.title('Top 5 most viewed artist on Youtube')
sns.barplot(data = artists_youtube, x = artists_youtube.reset_index()["Artist"], y =artists_youtube.reset_index()["Views"], palette = 'Reds_r')
plt.show()
```

**Most streamed artists on Spotify**

```python
#Filter top 3 artist on Spotify

artists_spotify = data.groupby('Artist').sum()['Stream'].sort_values(ascending=False).head(5)
artists_spotify = artists_spotify.sort_values(ascending = False)
artists_spotify = pd.DataFrame(artists_spotify)
```

```python
artists_spotify.reset_index()["Artist"]
```

```python
plt.figure(figsize=(19,6))
plt.title('Top 5 most streamed artist on Spotify')
sns.barplot(data = artists_spotify, x = artists_spotify.reset_index()["Artist"], y = artists_spotify.reset_index()["Stream"], palette = 'Greens_r')
plt.ylabel('Billion of streams')
plt.show()
```

*The lists of the most listened-to artists on YouTube and Spotify may not be the same due to each platform having its own recommendation and content discovery system, which can lead to the promotion of different artists. Additionally, the type of content found on each platform may be different, with YouTube allowing users to upload home videos, while Spotify focuses on high-quality audio playback. Furthermore, the audience on each platform may also be different, with certain artists appealing to specific audiences on each platform. In summary, there are multiple factors that can influence the list of the most listened-to artists on each platform, which can lead to differences between them.*

```python
top3songsonYT = data[['Track','Views','Danceability','Energy']].sort_values('Views', ascending=False).head(5)
top3songsonYT.head()
```

```python
sns.catplot(x='Track', y='Value', hue='Variable', data=top3songsonSptf.melt(id_vars=['Track', 'Stream'], var_name='Variable', value_name='Value'), kind='bar', palette='Greens_r', aspect = 2)
plt.title("Danceability and Energy of top Songs by Streams on Spotify")
plt.show()
```

*Although they are not the same songs, the patterns of the most listened lists on both platforms are very similar. The most listened song on both platforms transmits a lot of energy, while the second one is more danceable.*

*This may be because on both platforms, users are mainly looking for songs that make them feel good and allow them to enjoy music in different ways. It is possible that these two characteristics, energy and dance, are the main factors influencing the choice of the most listened songs*

**Videos with the highest like ratios**


```python
data['like_percent'] = data['Likes']/ data['Views']
data['like_percent'] = data['like_percent'].apply(lambda x: round(x, 3))
data2 = data[data['Views'] > 5000000]
data2 = data2.sort_values(by = 'like_percent', ascending = False).head(5)[['Artist', 'Track', 'Views', 'like_percent']]
```

```python
data2
```

```python
plt.figure(figsize = (12,5))
sns.barplot(data2, x = data2['Track'], y = data2['like_percent'], palette = 'Reds_r')
plt.title('Top 5 songs with best like/view ratio')
plt.xlabel('Song', color = 'gray')
plt.ylabel('Like/view', color = 'gray')
plt.show()
```

*There are several factors that can influence why the most played songs do not have the highest view/like ratio.*

*One of them is that the number of plays can be influenced by external factors, such as promotion on radio, TV, online advertising, social media, among others. This can lead to a song being played many times, even if users don't like the song as much or don't identify with it.*

*On the other hand, the number of "likes" a song receives can be more influenced by the content of the song itself, such as the lyrics, melody, rhythm, singer's voice, musical style, etc. It's possible that a song that is not promoted as much, but has great musical content, generates many likes and is highly appreciated by users, but doesn't have as many plays.*

*Additionally, the usage dynamics of both platforms can be different: while it is more common to listen to complete songs on YouTube, on Spotify users can skip between songs and create personalized playlists, which can influence the number of plays and likes for each song.*

*In summary, the view/like ratio can be influenced by several factors, so the most played songs may not necessarily be the most appreciated in terms of personal tastes.*

**Conclusions**



1.   The number of plays a song receives can be influenced by external factors such as promotion, which can lead to high play counts even if users don't necessarily like the song.
2.  The number of likes a song receives is more influenced by the content of the song itself, such as the lyrics, melody, rhythm, and musical style.
3. The usage dynamics of platforms like YouTube and Spotify can differ, with users on Spotify often creating personalized playlists and skipping between songs, which can influence play counts and likes.
4. The most listened songs on YouTube and Spotify have similar patterns, with the most listened song on both platforms tending to be energetic and the second song more danceable.
5. Users on both platforms seem to prioritize songs that make them feel good and allow them to enjoy music in different ways, suggesting that energy and danceability are important factors in the choice of most listened songs.
