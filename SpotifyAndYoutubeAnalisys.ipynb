```python
#import libraries to use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the database
data = pd.read_csv('Spotify_Youtube.csv')
data.head()
```

```python
#Variable information 
data.info()
```

```python
#Delete categories
data = data.drop(['Title', 'Url_spotify', 'Uri', 'Description', 'official_video', 'Channel', 'Url_youtube'], axis = 1)
```

Group our dataset by album types to represnt share of views of each type




```python
streamperalbum=data.groupby('Album_type').sum().loc[:,['Views','Stream','Likes','Comments']]
```

```python
streamperalbum.plot.pie(y='Views',colors=["#991E1E", "#A54545","#EE7575"],autopct='%1.0f%%').legend(bbox_to_anchor=(1.5, 1))
plt.title('Views Per Album Type')
plt.show()
```

El gráfico de tarta representa la distribución de los diferentes tipos de lanzamientos musicales, con los álbumes contribuyendo a la mayoría de los lanzamientos con un 74,4%. Los singles representan el 22,1% y los recopilatorios el 3,5% restante. Esto indica que los álbumes son la forma más popular de publicación de música, seguidos de los singles y las recopilaciones.

Group our dataset by album types to represnt share of stream of each type


```python
streamperalbum.plot.pie(y='Stream',colors=["#1E9941", "#44C167","#68F18F"],autopct='%1.0f%%').legend(bbox_to_anchor=(1.5, 1))
plt.title('Streams Per Album Type')
plt.show()
```

Most viewed artists on Youtube platform

```python
artists_youtube = data.groupby('Artist').sum()['Views'].sort_values(ascending=False).head(5)
artists_youtube = artists_youtube.sort_values(ascending = False)
artists_youtube = pd.DataFrame(artists_youtube)
```

```python
artists_youtube.reset_index()["Artist"]
```

```python
plt.figure(figsize=(19,6))
plt.title('Top 5 most viewed artist on Youtube')
sns.barplot(data = artists_youtube, x = artists_youtube.reset_index()["Artist"], y =artists_youtube.reset_index()["Views"], palette = 'Reds_r')
plt.show()
```

```python
top3songsonYT = data[['Track','Views','Danceability','Energy']].sort_values('Views', ascending=False).head(5)
top3songsonYT.head()
```

```python
sns.catplot(x='Track', y='Value', hue='Variable', data=top3songsonYT.melt(id_vars=['Track', 'Views'], var_name='Variable', value_name='Value'), kind='bar', palette='Reds_r', aspect = 2)
plt.title("Danceability and Energy of top Songs by Views on Youtube")
plt.show()
```

```python
top3songsonSptf = data[['Track','Stream','Danceability','Energy']].sort_values('Stream', ascending=False).head(3)
top3songsonSptf.head()
```

```python
sns.catplot(x='Track', y='Value', hue='Variable', data=top3songsonSptf.melt(id_vars=['Track', 'Stream'], var_name='Variable', value_name='Value'), kind='bar', palette='Greens_r', aspect = 2)
plt.title("Danceability and Energy of top Songs by Streams on Spotify")
plt.show()
```

Videos with the highest like ratios


```python
data['like_percent'] = data['Likes']/ data['Views']
data['like_percent'] = data['like_percent'].apply(lambda x: round(x, 3))
data2 = data[data['Views'] > 5000000]
data2 = data2.sort_values(by = 'like_percent', ascending = False).head(5)[['Artist', 'Track', 'Views', 'like_percent']]
```

```python
data2
```

```python
plt.figure(figsize = (12,5))
sns.barplot(data2, x = data2['Track'], y = data2['like_percent'], palette = 'Reds_r')
plt.title('Top 5 songs with best like/view ratio')
plt.xlabel('Song', color = 'gray')
plt.ylabel('Like/view', color = 'gray')
plt.show()
```
